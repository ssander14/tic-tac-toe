```python
#Question 1
import requests
import json

lat = "42.098701"
lon = "-75.912537"
lat1 = "40.730610"
lon1 = "-73.935242"

def return_address(lat,lon):
    weather = requests.get(f"https://api.weather.gov/points/{lat},{lon}")
    json_file = weather.json()
    city = json_file['properties']['relativeLocation']['properties']['city']
    state = json_file['properties']['relativeLocation']['properties']['state']
    print(f"Address is {city}, {state}.")

return_address(lat,lon)
return_address(lat1,lon1)
```

```python
#Question 2-5
country_code = "US"
country = requests.get(f"https://date.nager.at/api/v3/IsTodayPublicHoliday/{country_code}")
print(country)
if country == "<Response [200]>":
    print("Today is a public holiday")
else:
    print("Today is not a public holiday")
#  This uses a country code to check if today is a public holiday. If returns 204
#  it is not a public holiday but if it returns 200 it is a public holiday.
```

```python
#Question 4
# This API returns a random breaking bad quote or quotes along with the character
# or characters who said it. It does not require payment or a key to use.

# Thge base url is "https://api.breakingbadquotes.xyz/v1/quotes" and you can use it
# two different ways

# The first way you use is justy by using the base url and it returns on quote
# and who ever said it
breaking_bad_quote = requests.get("https://api.breakingbadquotes.xyz/v1/quotes")
breaking_bad_quote_json = breaking_bad_quote.json()
print(breaking_bad_quote_json)

# The first way you use is justy by using the base url and then adding a slash and
# the number of quotes you would like. It then returns the number of quotes you wanted
# and who ever said each quote
number_of_quotes = "3"
breaking_bad_quotes = requests.get(f"https://api.breakingbadquotes.xyz/v1/quotes/{number_of_quotes}")
breaking_bad_quotes_json = breaking_bad_quotes.json()
print(breaking_bad_quotes_json)
```
